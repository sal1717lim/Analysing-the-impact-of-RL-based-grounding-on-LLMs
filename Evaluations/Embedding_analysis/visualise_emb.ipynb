```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
first_embeddings=np.load("ftfirst_embeddings.npy")
mean_embeddings=np.load("ftmean_embeddings.npy")
first_embeddings2D=np.load("ftfirst_embeddings2D.npy")
mean_embeddings2D=np.load("ftmean_embeddings2d.npy")
```

```python
for i in range(0,6000,1000):
    plt.scatter(mean_embeddings2D[i:i+1000,0],mean_embeddings2D[i:i+1000,1],label="prompt"+str(i/1000))
plt.legend()
    
```

```python
p1_to_Glam_mean=(mean_embeddings[1000:2000,:]-mean_embeddings[0:1000,:]).mean(axis=0)
p2_to_Glam_mean=(mean_embeddings[2000:3000,:]-mean_embeddings[0:1000,:]).mean(axis=0)
p3_to_Glam_mean=(mean_embeddings[3000:4000,:]-mean_embeddings[0:1000,:]).mean(axis=0)
p4_to_Glam_mean=(mean_embeddings[4000:5000,:]-mean_embeddings[0:1000,:]).mean(axis=0)
p6_to_Glam_mean=(mean_embeddings[5000:6000,:]-mean_embeddings[0:1000,:]).mean(axis=0)
np.save("mean_drift.npy",np.concatenate([[p1_to_Glam_mean,p2_to_Glam_mean,p3_to_Glam_mean,p4_to_Glam_mean,p6_to_Glam_mean]]))
```

```python
p1_to_Glam_first=(first_embeddings[1000:2000,:]-first_embeddings[0:1000,:]).mean(axis=0)
p2_to_Glam_first=(first_embeddings[2000:3000,:]-first_embeddings[0:1000,:]).mean(axis=0)
p3_to_Glam_first=(first_embeddings[3000:4000,:]-first_embeddings[0:1000,:]).mean(axis=0)
p4_to_Glam_first=(first_embeddings[4000:5000,:]-first_embeddings[0:1000,:]).mean(axis=0)
p6_to_Glam_first=(first_embeddings[5000:6000,:]-first_embeddings[0:1000,:]).mean(axis=0)
np.save("first_drift.npy",np.concatenate([[p1_to_Glam_first,p2_to_Glam_first,p3_to_Glam_first,p4_to_Glam_first,p6_to_Glam_first]]))
```
